# **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Read Dataset**

```python
data = pd.read_csv('/content/Sales.csv')
```

# **Explore Data**

```python
data.head()
```

# **Plot**

```python
plt.plot(data['Quantity'],data['Unit price'],color='red' , marker = 'o' , label= 'Unit Price ')
plt.plot(data['Quantity'],data['gross income'],color='blue' , marker = 'o' , label= 'Gross Income ')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.title('Unit Price Per Quantity')
plt.style.use('dark_background')
plt.legend()
```

# **Scatter**

```python
plt.scatter(data['cogs'],data['Rating'] , color = 'blue')
plt.scatter(data['Unit price'],data['Rating'] , color = 'red')
```

# **Pie**

```python
plt.pie(data['gross income'],autopct='%1.1f%%')
plt.title('Gross Income')
plt.axis=['equal']
```

# **Bar**

```python
plt.bar(data['Quantity'],data['Unit price'])
plt.text(6,-12,'Quantity Per Price')
```

# **Hisogram**

```python
plt.hist(data['gross income'])
plt.title('Gross Income')
plt.axis=['equal']
```

# **BoxPlot**

```python
sns.boxplot(data['Rating'])
```
